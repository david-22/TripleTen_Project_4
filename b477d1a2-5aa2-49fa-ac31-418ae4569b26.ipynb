### [C1] ¿Cuántos artículos compran normalmente las personas en un pedido? ¿Cómo es la distribución?

````python
items_buyer_average = instacart_orders.merge(order_products, on='order_id')
items_buyer_average = items_buyer_average.groupby(['user_id', 'order_id']).agg({'product_id': 'count'}).reset_index()
items_buyer_average.info()
items_buyer_average.rename(columns={'product_id': 'items_purchased'}, inplace=True)
items_buyer_average = items_buyer_average.groupby('user_id').agg({'order_id': 'count', 'items_purchased': 'sum'}).reset_index()

def average(row):
    return (row['items_purchased']/row['order_id'])

items_buyer_average['average'] = items_buyer_average.apply(average, axis=1)
items_buyer_average['average'] = items_buyer_average['average'].round(0)
items_buyer_average = items_buyer_average.sort_values(by= 'average',ascending=False)

items_buyer_average = items_buyer_average.groupby('average')['order_id'].count().reset_index()
items_buyer_average.rename(columns={'order_id': 'no_clients'},inplace=True)

items_buyer_average.plot(kind='bar', title='Articulos promedio comprados por usuario', xlabel='Cantidad de articulos', ylabel='Cantidad de personas')
plt.show()
````

### [C3] Para cada producto, ¿cuál es la proporción de las veces que se pide y que se vuelve a pedir?

````python
merged = instacart_orders.merge(order_products, on='order_id')
total_orders = merged.groupby('product_id')['order_id'].count()
repeated_orders = merged[merged['reordered'] == '1'].groupby('product_id')['order_id'].count()

repeat_rate = ((repeated_orders/total_orders)*100).fillna(0)
repeat_rate.columns = ['product_id', 'repeat_rate']
repeat_rate = repeat_rate.head(20)
repeat_rate.plot(kind='bar', title = 'Tasa de repeticion por producto')
plt.show()
````
